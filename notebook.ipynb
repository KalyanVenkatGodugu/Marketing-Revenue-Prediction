```python
import pandas as pd

df = pd.read_csv("marketing_and_product_performance.csv")

# Remove ID columns
df = df.drop(columns=['Campaign_ID', 'Product_ID', 'Customer_ID'], errors='ignore')

df.head()
```

```python
df = df[['Budget', 'Clicks', 'Conversions', 'ROI', 'Revenue_Generated']]

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df.drop('Revenue_Generated', axis=1)
y = df['Revenue_Generated']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test, pred))
print("R2 Score:", r2_score(y_test, pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, pred)
plt.xlabel("Actual Revenue")
plt.ylabel("Predicted Revenue")
plt.title("Actual vs Predicted Revenue")
plt.show()
```

```python
import pandas as pd

importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

importance.sort_values(by='Coefficient', ascending=False)
```
